```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
```

```python
import pickle
with open('/content/drive/MyDrive/am5t2/pickle/credit.pkl', 'rb') as f:
  X_credit_treinamento, Y_credit_treinamento, X_credit_teste, Y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, X_credit_teste.shape
```

```python
Y_credit_treinamento.shape, Y_credit_teste.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore_credit = DecisionTreeClassifier(criterion='entropy', random_state=0)
arvore_credit.fit(X_credit_treinamento, Y_credit_treinamento)
```

```python
pa_arvore= arvore_credit.predict(X_credit_teste)
pa_arvore
```

```python
Y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
pa = accuracy_score(Y_credit_teste, pa_arvore)
pa
```

```python
from yellowbrick.classifier import ConfusionMatrix
cmt = ConfusionMatrix(arvore_credit)
cmt.fit(X_credit_treinamento, Y_credit_treinamento)
cmt.score(X_credit_teste, Y_credit_teste)
```

```python
print(classification_report(Y_credit_teste, pa_arvore))
```

```python
from sklearn import tree
pa_arvore = ['income', 'age', 'loan']
figura, axes = plt.subplots(nrows=1, ncols=1, figsize=(20,20))
tree.plot_tree(arvore_credit, feature_names=pa_arvore, class_names=['0','1'], filled=True)
figura.savefig('/content/drive/MyDrive/am5t2/treemap.png')
```

# Precisão do Algoritmo de Arvore

```python
devedores = np.sum(Y_credit_teste == 1)
```

```python
pa_a=pa
pb_a=devedores/len(Y_credit_teste)

pa_a, pb_a
```

```python
pab_a=(pa_a * pb_a)/((pa_a* pb_a) + ((1-pa_a)*(1-pb_a)))
print(f'P(AB) = {pab_a *100:.2f}%')
```

# Precisão do Algoritmo de Random Forest

```python
rpb=devedores/len(Y_credit_teste)

rpa, rpb
```

```python
pab_r=(rpa * rpb)/((rpa* rpb) + ((1-rpa)*(1-rpb)))
print(f'P(AB) = {pab_r *100:.2f}%')
```
